```python
import pandas as pd
import matplotlib.pyplot as plt

s_frame = pd.read_csv("GME_stock.csv")
print('s_frame(row, col) =', s_frame.shape)
s_frame.head(5) # Print the first 9 entries.
```

```python
data = s_frame[['date', 'low_price', 'high_price', 'open_price', 'close_price', 'volume']]

data['date'] = pd.DatetimeIndex(data['date'])

plt.figure(figsize=(10, 8), constrained_layout=True)
c = ['green', 'red']
y = ['open_price', 'close_price']
l = 2.0

for j, c in zip(y, c):
  plt.plot(data['date'], data[j], color=c, linewidth=l, label=f'Daily {j}')
  plt.title('Gamestop Stock')
  plt.xlabel('Date')
  plt.ylabel('Stock Price')
  plt.legend()
  l -=0.5

plt.show()
```
